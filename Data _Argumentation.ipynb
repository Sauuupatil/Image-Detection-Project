```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_gen = ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.20,
    height_shift_range=0.20,
    shear_range=0.20,
    zoom_range=0.20,
    fill_mode='nearest',
    horizontal_flip=True,
    vertical_flip=True
)
```

```python
img
```

```python
img.shape
```

```python
img_reshape=img.reshape(1,408,612,3)
```

```python
img.shape
```

```python
img_reshape.shape
```

```python
c = 0

for batch in img_gen.flow(img_reshape,batch_size=32,
             save_to_dir = "C:/Users/saurabh/Pictures/rabbit1/",
             save_format= "jpg",
             save_prefix="rabbit"):
    c=c+1
    if c> 50:
        break
```
